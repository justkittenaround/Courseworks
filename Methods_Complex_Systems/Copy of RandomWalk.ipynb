```
x = [0.5,0.5]
```

```
n = 100
```

```
X = np.zeros((n,2)) 
```

```
X[0,:] = x
```

```
X
```

```
def get_rand():

    r = 2*(np.random.random_sample((1, 2))) - 1

    return r
```

```
get_rand()
```

```
X[0,:]
```

```
X[0,:] + get_rand()
```

```
for i in range(n - 1):

   X[i+1,:] = X[i,:] + get_rand()
```

```
X
```

```
plt.plot(X[:,0],X[:,1],'-.', markersize=5, marker='^')
```
